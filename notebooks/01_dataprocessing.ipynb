```python
import pandas as pd
import numpy as np

def load_data(filepath):
    df = pd.read_csv(filepath)

    if 'year' in df.columns:
        df['date'] = pd.to_datetime(df[['year','month','day','hour']])
        df.set_index('date', inplace=True)
        df.drop(['No','year','month','day','hour'], axis=1, inplace=True)

    df = df.select_dtypes(exclude=['object'])
    df['pm2.5'] = df['pm2.5'].interpolate(method='linear')
    df = df.dropna()
    return df

def create_lags(df, target='pm2.5', n_lags=24):
    df_lag = df.copy()
    for i in range(1, n_lags + 1):
        df_lag[f'lag_{i}'] = df_lag[target].shift(i)
    return df_lag.dropna()

path = "/content/pollution.csv"
df = load_data(path)
df_processed = create_lags(df)

df_processed.to_csv("/content/pm25_processed.csv")
print("Saved pm25_processed.csv")
```
